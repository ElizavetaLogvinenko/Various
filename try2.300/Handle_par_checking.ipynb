```python
import numpy as np
```

```python
def data_verificate(input_d, ver_d):
    result = []
    for i in range(0, len(input_d)):#единицы измерения!!!, округления для уменьшения веса файлов 
        #(тут как бы любое число параметров)
#         print(input_d[i])
        print(ver_d[i])
        x = (abs(input_d[i][1]*2.5**2 - ver_d[i][0]))#mu
        print('x={}'.format(x))
        y = (abs(input_d[i][2]*2.5**2 - ver_d[i][1]))#sigma
        print('y={}'.format(y))
        tmp_res = [x, y]
        result.append(np.array(tmp_res))
    return result
```

```python
def handler(verificate_data):
    global_result = []
    for i in range(0, 100):# поменять на все файлы из папки
        input_data = np.load('current_par_CM_' + str(i) + '.npy', allow_pickle=True)
        print(input_data[1])
        local_result = data_verificate(input_data[1], verificate_data)
        global_result.append(np.array(local_result))
    return global_result
```

```python
verification_data = []
verification_data.append(np.array([925.9259259,   462.962963]))
verification_data.append(np.array([848.7654321,   385.8024691]))
verification_data.append(np.array([48.61111111,  11.11111111]))
verification_data.append(np.array([0.89, 0.1]))
```

```python
res = handler(verification_data)
# for i in res:
#     print(i)
```
