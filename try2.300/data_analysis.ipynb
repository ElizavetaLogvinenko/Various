```python
import numpy as np
```

```python
import pandas as pd
```

```python
with open('params.txt') as params:
    par_data = params.read()
```

```python
par_data
```

```python
par_data = par_data.split('\n')
```

```python
par_data
```

```python
data_array = []

for i in range(len(par_data)-1):
    cur_par = par_data[i].split(',')
    cur_par[0] = cur_par[0][1:]
    cur_par[-1] = cur_par[-1][:-1]
    for j in range(len(cur_par)):
#         print(cur_par)
        cur_par[j] = np.float(cur_par[j])
    data_array.append(cur_par)
```
